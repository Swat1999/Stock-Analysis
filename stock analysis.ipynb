# Moving Average

### So, now we have to see if there is any noise in the data and if there is, we will try to smoothen it out using the moving average to see if we can see some resistance level in the data.

```python
netflix['Open'].plot(label='No Moving average', figsize=(16,8))
netflix['MA-100'] = netflix['Open'].rolling(100).mean()
netflix['MA-100'].plot(label='Moving average of 100')
netflix['MA-200'] = netflix['Open'].rolling(200).mean()
netflix['MA-200'].plot(label='Moving average of 200')
plt.legend()
plt.title('Moving averages of 100 and 200 (Netflix)', size=18)
plt.xlabel('Years', size=16)
plt.ylabel('Open prices', size=16)
plt.xticks(size=18)
plt.yticks(size=18)
```

```python
starbucks['Open'].plot(label='No Moving average', figsize=(16,8))
starbucks['MA-100'] = starbucks['Open'].rolling(100).mean()
starbucks['MA-100'].plot(label='Moving average of 100')
starbucks['MA-200'] = starbucks['Open'].rolling(200).mean()
starbucks['MA-200'].plot(label='Moving average of 200')
plt.legend()
plt.title('Moving averages of 100 and 200 (Starbucks)', size=18)
plt.xlabel('Years', size=16)
plt.ylabel('Open prices', size=16)
plt.xticks(size=14)
plt.yticks(size=14)
```

```python
amazon['Open'].plot(label='No Moving average', figsize=(16,8))
amazon['MA-100'] = amazon['Open'].rolling(100).mean()
amazon['MA-100'].plot(label='Moving average of 100')
amazon['MA-200'] = amazon['Open'].rolling(200).mean()
amazon['MA-200'].plot(label='Moving average of 200')
plt.legend()
plt.title('Moving averages of 100 and 200 (amazon)',size=18)
plt.xlabel('Years', size=16)
plt.ylabel('Open prices', size=16)
plt.xticks(size=14)
plt.yticks(size=14)
```

### we should be keeping in mind that, having too little moving average will create noisy data and too large moving average will make the data inaccurate, so there should be a balanced moving average that won't show any noise.

### Here in this case
### 1. Both starbucks and netflix despite having the same moving average is still showing noisy hence we need a              much bigger value to balance it out.
### 2. In case of amazon, a 'moving average of 100' is making it noisy but a 'moving average of 200' is somewhat                  less noisy compared to Starbucks and netflix.

### Recommendation: If we can use a moving average value between 200-300, perhaps it will even it out.

# Financial Details:

## Percentage of Change

### This method will help us predict the future change in percentage keeping in mind the volatility of this stocks

```python
netflix['returns'] = (netflix['Close']/netflix['Close'].shift(1)) - 1
starbucks['returns'] = (starbucks['Close']/starbucks['Close'].shift(1)) - 1
amazon['returns'] = (amazon['Close']/amazon['Close'].shift(1)) - 1
```

```python
netflix
```

```python
starbucks
```

```python
amazon
```

### Now that we have gotten our returns, we will now plot the volatility of these returns on a histogram and see which stock is most volatile and it is the one we should be avoiding to buy/invest

```python
plt.figure(figsize=(20,10))
netflix['returns'].hist(bins=100, label='Netflix', alpha=0.6)
starbucks['returns'].hist(bins=100, label='Starbucks', alpha=0.6)
amazon['returns'].hist(bins=100, label='Amazon', alpha=0.6)
plt.title('Volatility of Netflix, Starbucks and Amazon', size=16)
plt.legend()
plt.xlabel('Returns', size=16)
plt.ylabel('Range', size=16)
plt.xticks(size=14)
plt.yticks(size=14)
```

### key point: The more wider the graph, the more volatile the stocks is

### So what we can observe from this burj khalifa graph here is that, Netflix and Amazon stocks are more volatile hence they are more risky to buy or invest. However, starbucks is on a safe side as its less volatile and is giving more profit as we can see the range going up

### Lets understand it more using Kernel Density (KDE).

## KDE Plot 

```python
plt.figure(figsize=(20,10))
netflix['returns'].plot(kind='kde', label='Netflix')
starbucks['returns'].plot(kind='kde', label='Subscriber')
amazon['returns'].plot(kind='kde', label='Amazon')
plt.legend()
plt.xlabel('Returns', size=18)
plt.ylabel('Range', size=18)
plt.xticks(size=16)
plt.yticks(size=16)
plt.title('KDE Representation of Volatility', size=20)
```

### This is a much better representation as it is normalized. We can see the change from the histogram to here.
### It remains the same, Starbucks is less volatile compared to Amazon and Netflix

## Cumulative Return

```python
netflix['cumulative returns'] = (1 + netflix['returns']).cumprod()
starbucks['cumulative returns'] = (1 + starbucks['returns']).cumprod()
amazon['cumulative returns'] = (1 + amazon['returns']).cumprod()
```

```python
netflix
```

```python
starbucks
```

```python
amazon
```

```python
netflix['cumulative returns'].plot(label = 'Netflix', figsize=(16,8))
starbucks['cumulative returns'].plot(label = 'Starbucks')
amazon['cumulative returns'].plot(label = 'Amazon')
plt.legend()
plt.title('Cumulative Returns of Netflix, Starbucks and Amazon', size=18)
plt.xticks(size=14)
plt.yticks(size=14)
plt.xlabel('Years', size=16)
plt.ylabel('Cumulative Returns', size=16)
```

### So what does this Cumulative returns means,
### It means what is the best daily return for the minimum amount of money invested on the stock.
### So what do we understand from this graph, is that Amazon though its volatile and has noises, gave excellent returns on investment compared to Netflix and starbucks.

### So how can it give more returns if its volatile?
#### Answer: Its simple, the more the risk involved in the stock, the more return it gives. If you want to play safe then Netflix or Starbucks is for you however, they give low returns
